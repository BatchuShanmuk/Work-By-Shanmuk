```python
import pandas as pd

# Creating a sample retail sales dataset
data = {
    "Date": [
        "2023-01-01","2023-02-01","2023-03-01","2023-04-01","2023-05-01",
        "2023-06-01","2023-07-01","2023-08-01","2023-09-01","2023-10-01"
    ],
    "Product": [
        "Laptop","Mobile","Tablet","Laptop","Mobile",
        "Tablet","Laptop","Mobile","Tablet","Laptop"
    ],
    "Sales": [50000, 40000, 30000, 52000, 41000, 32000, 60000, 45000, 35000, 65000],
    "Profit": [8000, 6000, 4000, 8500, 6200, 4500, 10000, 7000, 5000, 11000]
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("retail_sales.csv", index=False)

df
```

```python
import matplotlib.pyplot as plt
```

```python
top_products = df.groupby("Product")["Sales"].sum()

plt.figure(figsize=(8,5))
top_products.plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
monthly_sales = df.groupby(df["Date"].dt.month)["Sales"].sum()

plt.figure(figsize=(8,5))
plt.plot(monthly_sales)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Profit"], bins=5)
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

### Insights

1. The bar chart shows that certain products generate significantly higher sales compared to others.
2. The line chart highlights sales trends over time, helping identify peak and low sales periods.
3. The histogram and scatter plot reveal profit distribution and a positive correlation between sales and profit.
